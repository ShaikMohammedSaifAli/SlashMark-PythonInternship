```python
import matplotlib.pyplot as plt
import cv2
import easyocr
from IPython.display import Image
```

```python
Image("F:\INTERNSHIP\SLASH MARK\Advanced Project\Plates\ROI.png")
```

```python

reader = easyocr.Reader(['en'])
```

```python

output = reader.readtext('F:\INTERNSHIP\SLASH MARK\Advanced Project\Plates\ROI.png')
output
```

```python
cord = output[-1][0]
cord
```

```python
a = list(zip(*cord))
a
```

```python
min(a[0])
```

```python
min(a[1])
```

```python
max(a[0])
```

```python

x_min, y_min = [int(min(idx)) for idx in zip(*cord)]
x_min, y_min
```

```python

x_max, y_max = [int(max(idx)) for idx in zip(*cord)]
x_max, y_max
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 20, 30
```

```python
image = cv2.imread('F:\INTERNSHIP\SLASH MARK\Advanced Project\Plates\ROI.png')
cv2.rectangle(image,(x_min,y_min),(x_max,y_max),(0,0,255),2)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
